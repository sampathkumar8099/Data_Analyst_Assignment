```python
def charges(x,y,z):
    if x == 'a':
        if y == 'Forward charges':
            return (29.5+(z-0.25) * 23.6)
        elif y == "Forward and RTO charges":
            return (29.5+(z-0.25)*23.6+13.6+23.6)
    elif x == "b":
        if y == 'Forward charges':
            return (33.0+(z-0.50) * 28.3)
        elif y == "Forward and RTO charges":
            return (33.0+(z-0.50)*28.3+20.5+28.3)
    elif x == "c":
        if y == 'Forward charges':
            return (40.1+(z-0.75) * 38.9)
        elif y == "Forward and RTO charges":
            return (40.1+(z-0.75)*38.9+31.9+38.9)
    elif x == "d":
        if y == 'Forward charges':
            return (45.4+(z-1.25) * 44.8)
        elif y == "Forward and RTO charges":
            return (45.5+(z-1.25)*44.8+41.3+44.8)
    else:
        if y == 'Forward charges':
            return (56.6+(z-1.75) * 55.5)
        elif y == "Forward and RTO charges":
            return (56.6+(z-1.75)*55.5+50.7+55.5)
```
